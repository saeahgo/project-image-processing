```python
# (b)
import skimage
import matplotlib.pyplot as plt
import numpy as np

# Load the cameraman test image
cameraman_image = skimage.data.camera()

# Display the image in a new figure
plt.figure()
plt.imshow(cameraman_image, cmap="gray", vmin=0, vmax=255) # Display the image as grayscale with a minimum of 0 and maximum of 255
plt.title("Original Cameraman Image")
plt.axis("off")  # Remove axis tick marks
plt.colorbar() # Display the grayscale
```

(d) The resolution of the cameraman image in megapixels is 0.262144.
($\frac{512 \times 512}{1,000,000}=0.262144)$

```python
# (e)
# Decrease all pixel values by 50 using vectorized operations
modified_cameraman_image = cameraman_image - 50

# Display the modified image in a new figure
plt.figure()
plt.imshow(modified_cameraman_image, cmap="gray", vmin=0, vmax=255)
plt.title("Cameraman Modified")
plt.axis("off")
plt.colorbar()
```

(f) The  black color pixels changed to bright (or white) pixels in the modified cameraman image, compared to the original cameraman image. In the original cameraman image, the man's suit and hair was dark (black) color.
The result is a bit different from what I expected. Since 0 is taken to be black and 255 is taken to be white in the grayscale, with the modified cameraman image, I expected the black colors (i.e. pixel is close to 0) will become darker than the original, but it gets brighter.

```python
# (g)
def brighten(image, amount):
  image16bit = image.astype(np.int16) # 255 is not an upper limit
  img_modified = image16bit + amount
  img_modified[img_modified > 255] = 255 # If bigger than 255 make the value to 255
  img_modified[img_modified < 0] = 0 # If less than 0, then make the value to 0
  img_modified = img_modified.astype(np.uint8)

  return img_modified
```

```python
# (h)
brightened_5 = brighten(cameraman_image, 5)
brightened_10 = brighten(cameraman_image, 10)
brightened_50 = brighten(cameraman_image, 50)

fig, ax = plt.subplots(1, 3)

ax[0].imshow(brightened_5, cmap="gray")
ax[0].axis("off")
ax[0].set_title("Brightened 5")

ax[1].imshow(brightened_10, cmap="gray")
ax[1].axis("off")
ax[1].set_title("Brightened 10")

ax[2].imshow(brightened_50, cmap="gray")
ax[2].axis("off")
ax[2].set_title("Brightened 50")
```

```python
# (i)
# Convert to floating point and normalize
cameraman_float = cameraman_image.astype(np.float32)  # Convert to float
cameraman_normalized = (cameraman_float - cameraman_float.min()) / (cameraman_float.max() - cameraman_float.min())

# Display the normalized image
plt.imshow(cameraman_normalized, cmap="gray", vmin=0, vmax=1)
plt.title("Normalized Cameraman Image (0 to 1)")
plt.axis("off")
plt.colorbar()
```

Yes, the image (i) looks reasonable. The normalized image should look identical to the original image, normalization doesn't change the color or brightness, it only rescales them to a new range. Thus it is correct that the image (i) look similar to the original grayscale version.
